## Simulation #1


Un ballon de 24 cm de rayon, positionné à x = -6 m, est immobile devant une personne (représentée par un cylindre vertical de 1,7 m de hauteur et 40 cm de diamètre). Un mur se trouve à une distance de 10 m du ballon. Ce mur a une hauteur de 6 m, 1 m d’épaisseur et 10 m de long (sur z). Cette personne frappe le ballon avec une vitesse de 12 m/s avec un angle de 60 degrés. Le programme s’arrête lorsque le ballon touche le sol (après avoir rebondi sur le mur). 
Vous devez créer un sol dont la dimension en x est de 30 m. L’épaisseur du sol est de 10 cm et la dimension en z est de 10 m. 
Vous devez aussi compter le temps pour afficher le temps lorsque le ballon touche le mur. Vous devez aussi affichez le temps de vol de la trajectoire (à la fin du programme).
Utilisez l’attribut make_trail=True lorsque vous déclarez votre objet afin de suivre sa trajectoire par une trace, que VPython dessinera, au fur et à mesure que votre objet se déplace.

```python
from vpython import *

# =========================
# PARAMÈTRES DE LA SCÈNE
# =========================
scene.width = 1000
scene.height = 600
scene.center = vec(2, 2, 0)
scene.background = color.black

# =========================
# CONSTANTES
# =========================
g = 9.8
dt = 0.001

rayon_ballon = 0.24
vitesse_initiale = 12
angle_degres = 60
angle_radians = radians(angle_degres)

# Position initiale du ballon
x_ballon = -6
y_ballon = rayon_ballon
z_ballon = 0

# Le mur est à 10 m du ballon
x_face_gauche_mur = x_ballon + 10
epaisseur_mur = 1
hauteur_mur = 6
longueur_mur = 10

# =========================
# OBJETS
# =========================

# Sol
sol = box(
    pos=vec(0, -0.05, 0),
    size=vec(30, 0.1, 10),
    color=color.green
)

# Personne
personne = cylinder(
    pos=vec(x_ballon - 0.4, 0, 0),
    axis=vec(0, 1.7, 0),
    radius=0.2,
    color=color.blue
)

# Mur
mur = box(
    pos=vec(x_face_gauche_mur + epaisseur_mur/2, hauteur_mur/2, 0),
    size=vec(epaisseur_mur, hauteur_mur, longueur_mur),
    color=color.white
)

# Ballon
ballon = sphere(
    pos=vec(x_ballon, y_ballon, z_ballon),
    radius=rayon_ballon,
    color=color.orange,
    make_trail=True
)

# =========================
# VITESSE INITIALE
# =========================
vx = vitesse_initiale * cos(angle_radians)
vy = vitesse_initiale * sin(angle_radians)
vz = 0

v = vec(vx, vy, vz)

# =========================
# TEMPS
# =========================
t = 0
temps_mur = None
rebond_fait = False

# =========================
# BOUCLE PRINCIPALE
# =========================
while True:
    rate(1000)

    # Mise à jour de la vitesse
    v.y = v.y - g * dt

    # Mise à jour de la position
    ballon.pos = ballon.pos + v * dt

    # Mise à jour du temps
    t = t + dt

    # Collision avec le mur
    if rebond_fait == False:
        if ballon.pos.x + ballon.radius >= mur.pos.x - mur.size.x/2:
            if ballon.pos.y >= 0 and ballon.pos.y <= mur.pos.y + mur.size.y/2:
                if abs(ballon.pos.z - mur.pos.z) <= mur.size.z/2:
                    temps_mur = t
                    print("Temps lorsque le ballon touche le mur =", round(temps_mur, 3), "s")

                    # On replace le ballon juste devant le mur
                    ballon.pos.x = mur.pos.x - mur.size.x/2 - ballon.radius

                    # Rebond simple sur un mur vertical
                    v.x = -v.x

                    rebond_fait = True

    # Collision avec le sol
    if ballon.pos.y <= ballon.radius:
        print("Temps de vol total =", round(t, 3), "s")
        break
```
